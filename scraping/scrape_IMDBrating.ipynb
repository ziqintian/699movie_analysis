```python
def get_rating(movie_id):
    """Scrapes the rating for a single movie ID"""
    url = f"https://www.imdb.com/title/{movie_id}/"
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    print(soup)
    rating = soup.find("span", {"itemprop": "ratingValue"})
    print(rating)
    if rating:
        return float(rating.text.strip())
    else:
        return None

def get_ratings(movie_ids):
    """Scrapes the ratings for a list of movie IDs"""
    ratings = []
    for movie_id in movie_ids:
        rating = get_rating(movie_id)
        ratings.append({"movie_id": movie_id, "rating": rating})
    return pd.DataFrame(ratings)

# Example usage:
movie_ids = df['id'][:5]
ratings_df = get_ratings(movie_ids)
ratings_df
```

```python
df['id_nott'] = df['id'].str.replace('tt', '')
```

```python
from imdb import IMDb

# create an instance of the IMDb class
imdb = IMDb()

# loop through the movie names and get the movie IDs
movies = []
for movie_id in df['id_nott']:
    movie = imdb.get_movie(movie_id)
    rating = movie.get('rating')
    movies.append({"id": movie_id, "rating": rating})

movies
```
