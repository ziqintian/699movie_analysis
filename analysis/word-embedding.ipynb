```python
# Load the stop words from a text file
with open('data/stoplist.txt') as f:
    stop_words = f.read().splitlines()

# Define a function to remove stop words from text
def remove_stop_words(text):
    # Process the text with spaCy
    doc = nlp(text)
    # Remove stop words from the processed Doc object
    non_stop_words = [token.text for token in doc if token.text.lower() not in stop_words]
    # Join the remaining words into a string
    processed_text = ' '.join(non_stop_words)
    return processed_text
```

```python
# Apply the function to the 'Review' column of the DataFrame
df1['Processed_Review'] = df1['Review'].apply(remove_stop_words)
df1
```

```python
# Initialize an empty list to store the results
results = []

# Iterate over the rows in the DataFrame
for _, row in df5.iterrows():
    # Create a spaCy Doc object for the review text
    doc = nlp(row['Review'])
    
    # Initialize counters for verbs, nouns, and adjectives
    verb_count = 0
    noun_count = 0
    adj_count = 0
    adv_count = 0
    pron_count = 0
    word_count = 0
    
    # Iterate over the tokens in the Doc object
    for token in doc:
        word_count += 1
        if token.pos_ == 'VERB':
            verb_count += 1
        elif token.pos_ == 'NOUN':
            noun_count += 1
        elif token.pos_ == 'ADJ':
            adj_count += 1
        elif token.pos_ == "ADV":
            adv_count += 1
        elif token.pos_ == "PRON":
            pron_count += 1
    
    # Add the counts to the results list
    results.append({
        'Movie ID': row['MovieID'],
        'Total word count': word_count,
        'Verb Count': verb_count,
        'Noun Count': noun_count,
        'Adjective Count': adj_count,
        'Adverb Count': adv_count,
        'Pronoun Count': pron_count
    })

# Create a new DataFrame from the results list
counts_df5 = pd.DataFrame(results)
```

```python
import pandas as pd
import string
from collections import Counter

# Function to clean and count the words in a text
def count_words(text):
    # Convert the text to lowercase
    text = text.lower()

    # Remove punctuation marks and special characters
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Split the text into a list of words
    word_list = text.split()

    # Count the occurrences of each word
    word_count = Counter(word_list)

    # Get the top 20 most frequent words
    top_20_words = word_count.most_common(20)

    return dict(top_20_words)

# Apply the 'count_words' function to the 'text' column and create a new 'word_counts' column
df5['word_counts'] = df5['Processed_Review'].apply(count_words)

# Print the dataframe
df5
```
