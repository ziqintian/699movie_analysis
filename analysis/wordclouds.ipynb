```python
import pandas as pd

# read data into a dataframe
freq = pd.read_csv('top30freq.csv', index_col=0)

# convert dataframe to dictionary
data_dict = freq.to_dict()['Count']
data_dict
```

```python
freq = freq.drop(['movies'])
freq
```

```python
top30 = freq[:30]
# convert dataframe to dictionary
top30_dict = top30.to_dict()['Count']
top30_dict
```

```python
from wordcloud import WordCloud

# create word cloud
wordcloud = WordCloud(width=5000, height=3000, background_color='white', colormap='GnBu').generate_from_frequencies(top30_dict)

# display word cloud
import matplotlib.pyplot as plt

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
